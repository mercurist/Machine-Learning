```python
def isPrime(var):
    if var == 1:
        return False
    idx = 2
    flag = True
    while idx**2 <= var:
        if var % idx == 0:
            flag = False
            break
        idx += 1
    return flag

n = int(input())
count = 0
for i in range(1, n + 1):
    if isPrime(i):
        print(i, end=" ")
        count += 1

print("\nthere are ", count, " primes from 1 to ", n)
```
