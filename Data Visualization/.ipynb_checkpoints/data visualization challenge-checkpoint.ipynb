#### Data Visualization Challenge
Given a movies dataset, the task is to:
- Plot the length of movie title names on the X-axis
- Frequency Count on Y-axis (no of movies having x characters in their title)

```python
counts = dict() 
for each_title in titles:
    length = len(each_title)
    if counts.get(length) is None:
        counts[length] = 1
    else :
        counts[length] += 1
```

```python
counts
```

```python
counts.values() 
```

```python
plt.figure(figsize = (15, 5)) 
plt.scatter(counts.keys(), counts.values()) 
plt.xlabel("Movie title length")
plt.ylabel("Frequency")
plt.title("Movie data visualization problem")  
  
plt.show() 
```

```python
df.dropna(axis = 0, subset = ['imdb_score', 'duration'], inplace = True)  
```

```python
df.shape[0] 
```

```python
df.head(15) 
```

```python
duration = list(df['duration']) 
score = list(df['imdb_score'])
ratings = dict() 
i = 0
for each_duration in duration:
    if ratings.get(int(each_duration)) is None:
        ratings[int(each_duration)] = [score[i]]
    else :
        ratings[int(each_duration)].append(score[i])
        
    i += 1 
    
```

```python
ratings
```

```python
for key in ratings:
    sum = 0
    n = len (ratings[key])
    for rating in ratings[key] :
        sum += rating
    sum = sum / n
    ratings[key] = sum
ratings
```

```python
# plotting the average rating corresponding to various durations
plt.figure(figsize= (15, 5)) 
plt.scatter(ratings.keys(), ratings.values()) 
```
