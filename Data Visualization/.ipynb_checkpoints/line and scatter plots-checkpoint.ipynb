```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10) 
y1 = x**2
y2 = 2*x + 3
print(x) 
print(y1) 
print(y2) 
```

```python
plt.style.use("seaborn") 
# marker shows the points through which the line is passing!
plt.plot(x, y1,c= "orange", label= "oranges", marker ="o")   
# for plotting separately
# plt.show() 
plt.plot(x, y2,c= "blue", label="blueberry", linestyle="dashed") 
plt.legend() 
# In jupyter notebook, you can see the object without the following statement.
plt.show() 
```

```python
subjects = ["Maths", "Chemistry", "Physics", "Computer Science", "English"]
weightage = [25, 15, 15, 30, 15]
# the explode parameter highlights the Computer science subject separately and the value denotes separation distance
plt.pie(weightage, labels = subjects, explode = (0,0,0,0.1,0))  
plt.show() 
```

#### Normal Distributions and Histograms

```python
xsn = np.random.randn(100)   
xsn 
```

```python
sigma = 10
mean =  50
X = np.round(xsn * sigma + mean)
X 
```

```python
plt.hist(X)
plt.ylabel("Frequency of students")
plt.xlabel("Marks in Mathematics")
plt.show()
```

```python
sigma_phy = 5
mean_phy = 70
X_phy = np.round(xsn * sigma_phy + mean_phy)
# alpha denotes the opacity, to better visualise the overlap
plt.hist(X, alpha = 0.7, label = "Mathematics") 
plt.hist(X_phy, alpha = 0.7, label = "Physics")  
plt.style.use("seaborn")
plt.ylabel("Frequency of students")
plt.xlabel("Marks")
plt.legend() 
plt.show() 

```
