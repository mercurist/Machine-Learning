```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
X = np.arange(10)
Y = 25 - ((X-5)**2)
print(X, Y)
```

```python
plt.title("Search for local maxima")
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.style.use("seaborn")
```

```python
x = random.randint(0, 9)
lr = 0.1
error = []
plt.plot(X, Y)

for i in range(50):
    gradient = -2 * (x-5)
    delta_x = lr * gradient
#     The only difference is in the update rule
    x = x + delta_x
    y = 25 - ((x-5)**2)
    
    error.append(25 - y)
    plt.scatter(x, y)
    print(x)
    
```

```python
plt.plot(error)
```
