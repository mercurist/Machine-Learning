```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

 Given a function f(x), we want to find the value of x that minimises the f

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.show()
```

Now we want to find out the value of x that minimises the function f(x)

```python
# we start off with a random value for x and a step-size (lr) aka learning rate
x = 0
lr = 0.1
error = []

plt.plot(X,Y)

threshold = 1e-3
# define the gradient descent function
# lets take 50 steps in the downhill direction
for i in range(50):
    #     basic calculus : y = (x-5)^2 gives us y' = 2 * (x-5)
    gradient = 2 * (x-5)
    #     update rule
    delta_x = lr * gradient
    x = x - delta_x
    y = (x-5)**2
    error.append(y)
    plt.scatter(x, y)
    
    print(x)
    #     to stop early
    if abs(delta_x) <= threshold:
        print("stopping at", str(i), " iterations.")
        break
```

We observe here that our x is converging to 5 as the final value is very close to the actual value.
(4.996038591874287 ~ 5)
