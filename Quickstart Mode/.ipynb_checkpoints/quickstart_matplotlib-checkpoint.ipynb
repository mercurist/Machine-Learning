```python
import matplotlib.pyplot as plt
import numpy as np
```

x = np.array([1,2,3,4,5])
y = x**2 + 3
print(x, y)

```python
plt.style.use("seaborn")
# plt.style.use("dark_background")

plt.title("squaring function")
plt.xlabel("timestamp")
plt.ylabel("stonks")

# Plot is a continuous plotting
plt.plot(x, y, label="actual prices", c= "orange")
plt.plot(x, x**2, label="expected prices", c= "red")

# Using a scatter plot instead
# plt.scatter(x, y, label="actual prices", c= "orange")
# plt.scatter(x, x**2, label="expected prices", c= "red")

# Setting boundaries explicitly
# plt.xlim(0,10)
# plt.ylim(0,20)
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
# plt.style.use("dark_background")

plt.title("squaring function")
plt.xlabel("timestamp")
plt.ylabel("stonks")

# Sudden changes in stonks - INFLATION
y[4] = x[4]**2 - 10 

# Using a scatter plot instead
plt.scatter(x, y, label="actual prices", c= "orange")
plt.scatter(x, x**2, label="expected prices", c= "red")

plt.plot(x, y, label="actual prices", c= "orange")

# Setting boundaries explicitly
# plt.xlim(0,10)
# plt.ylim(0,20)
plt.legend()
plt.show()
```
