```python
import cv2
import matplotlib.pyplot as plt
```

- Note that img.shape returns a tuple of a type ( no_of_columns, no_of_rows, no_of_channels)

```python
img = cv2.imread("./Images/pikachu.png")
img.shape
```

```python
plt.imshow(img)
```

```python
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### Alternate
- Use cv2.imshow() to see the image in a python script

```python
plt.imshow(rgb_image)
```

```python
plt.axis("off")
plt.imshow(gray_image)
```

```python
# takes column parameters first and then row and then the channels
cropped_image = rgb_image[200 : 1000, 350 : 1200, :]
plt.imshow(cropped_image)
plt.show()
```

```python
cropped_image.shape
```

```python
total_pixels = cropped_image.shape[0] * cropped_image.shape[1]
total_pixels
```

```python
new_image = cv2.resize(cropped_image, (100, 100))
plt.imshow(new_image)
```
